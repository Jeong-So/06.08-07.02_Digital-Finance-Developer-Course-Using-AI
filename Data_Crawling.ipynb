Requests 모듈 (Params)

```python
import requests

URL = 'http://httpbin.org/get'

params = {'data1':'data1','data2':'data2'} # params에 data를 담아서 요청
response = requests.get(URL, params=params)

print(response.status_code)
print(response.text)
```

```python
import requests

URL = 'http://httpbin.org/get?data1=data1&data2=data2'
# get방식으로 params에 안담고 url에 직접 담아서도 가능

# params = {'data1':'data1','data2':'data2'}
response = requests.get(URL)

print(response.status_code)
print(response.text)
```

Requests 모듈 (Header)

```python
import requests

URL = 'http://httpbin.org/get'

headers = {'Content-Type':'application/json; charset=utf-8'}
response = requests.get(URL, headers = headers)

print(response.status_code)
print(response.text)
```

Requests 모듈 (BODY)

```python
import requests

URL = 'http://httpbin.org/post'

data = {'data1':'data1','data2':'data2'}
response = requests.post(URL, data=data)

print(response.status_code)
print(response.text)
```

CSS 셀렉터

1. 태크 셀렉터 - p, h1, h2
2. id 셀렉터 (id 속성 활용) - #titld
3. class 셀렉터 (class 속성 활용) - .content
4. 속성 셀렉터 (태그내 속성 활용) - a[href], a[target="_blank"] 
5. 후손 셀렉터 (해당 태그 내 포함되는 태그) - div p
6. 자식 셀렉터 (해당 태그 바로 안에 포함되는 태크) - div > p

```python
# 3  (정답)

# 뉴스 크롤링

import requests
from bs4 import BeautifulSoup

dt = '20220611'
page = 1

while True :
  URL = f'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=003&date={dt}&page={page}'
  headers = { 'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36',}
  response = requests.get(URL, headers=headers)
  soup = BeautifulSoup(response.text, 'html.parser')
  if str(page) != soup.select_one('div.paging strong').text :
    break
  for item in soup.select('div.list_body ul li') :
    news_url = item.select('dt')[-1].select_one('a').attrs['href']
    print(news_url)
    news_resp = requests.get(news_url, headers=headers)
    news_soup = BeautifulSoup(news_resp.text, 'html.parser')
    try :
      title = news_soup.select_one('h2.media_end_head_headline').text
      content = news_soup.select_one('div.newsct_article').text.strip().replace('\n',' ')
      print(title)
      print(content)
    except :
      print()

  page=page+1
```

```python
# match

import re

text = "orange! I love orange!"
result = re.match("orange", text)
print(result)
```

```python
# search

import re

text = "I like orange! I love orange!"
result = re.search("orange", text)
print(result)
```

```python
# findall

import re

text = "I like orange! I love orange!"
result = re.findall("orange", text)
print(result)
```

```python
# finditer

import re

text = "I like orange! I love orange!"
result = re.finditer("orange", text)

for each in result:
  print(each)
```
